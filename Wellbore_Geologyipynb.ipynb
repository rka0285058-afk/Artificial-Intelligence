```python
import pandas as pd
import os

train_directory_path = '/content/rogii-wellbore-geology-prediction/train'

# Group files by well ID
well_data = {}
for filename in os.listdir(train_directory_path):
    if filename.endswith('.csv'):
        # Extract well ID (e.g., 'c99f9fae' from 'c99f9fae__horizontal_well.csv')
        well_id = filename.split('__')[0]
        if well_id not in well_data:
            well_data[well_id] = {'horizontal_well': None, 'typewell': None}

        if '__horizontal_well.csv' in filename:
            well_data[well_id]['horizontal_well'] = os.path.join(train_directory_path, filename)
        elif '__typewell.csv' in filename:
            well_data[well_id]['typewell'] = os.path.join(train_directory_path, filename)


# Merge data for each well
merged_dataframes = []
missing_files_count = 0

print("Merging data for each well...")
for well_id, files in well_data.items():
    if files['horizontal_well'] and files['typewell']:
        df_horizontal = pd.read_csv(files['horizontal_well'])
        df_typewell = pd.read_csv(files['typewell'])

        # Ensure 'TVT' column is present for merging, rename 'TVT_input' to 'TVT' if needed
        # Based on previous inspection, 'TVT' is already the primary TVD column.
        # This line is removed as 'TVT_input' had many NaNs and 'TVT' is complete.
        # if 'TVT_input' in df_horizontal.columns and 'TVT' not in df_horizontal.columns:
        #     df_horizontal.rename(columns={'TVT_input': 'TVT'}, inplace=True)

        # Sort both dataframes by TVT for merge_asof
        df_horizontal.sort_values(by='TVT', inplace=True)
        df_typewell.sort_values(by='TVT', inplace=True)

        # Use merge_asof to merge on the nearest TVT for geology labels
        # We only need 'TVT' and 'Geology' from df_typewell for the merge, other columns are features.
        merged_df = pd.merge_asof(
            df_horizontal,
            df_typewell[['TVT', 'Geology']],
            on='TVT',
            direction='nearest', # find the closest TVT in df_typewell
            suffixes=('_well', '_geology') # Suffixes are not strictly needed here as we drop GR from df_typewell
        )

        merged_df['well_id'] = well_id # Add well_id column for identification
        merged_dataframes.append(merged_df)
    else:
        missing_files_count += 1
        # print(f"Skipping well {well_id} due to missing horizontal_well or typewell file.")

print(f"Successfully merged data for {len(merged_dataframes)} wells.")
if missing_files_count > 0:
    print(f"Skipped {missing_files_count} wells due to missing files.")

# Concatenate all merged dataframes into a single training DataFrame
if merged_dataframes:
    df_train = pd.concat(merged_dataframes, ignore_index=True)
    print("\nCombined training DataFrame shape:", df_train.shape)
    print("\nCombined training DataFrame head:")
    print(df_train.head())
    print("\nCombined training DataFrame info:")
    df_train.info()
    print("\nUnique geology classes:", df_train['Geology'].unique())
else:
    print("No dataframes were merged.")
```

```python
print(f"Number of rows with missing Geology labels: {df_train['Geology'].isnull().sum()}")
print(f"Percentage of rows with missing Geology labels: {(df_train['Geology'].isnull().sum() / len(df_train)) * 100:.2f}%")

# Drop rows where 'Geology' is NaN, as these cannot be used for classification
df_train_cleaned = df_train.dropna(subset=['Geology']).copy()

print(f"\nShape of DataFrame after dropping NaN Geology rows: {df_train_cleaned.shape}")
print(f"Unique geology classes after cleaning: {df_train_cleaned['Geology'].unique()}")

# Also check for NaNs in feature columns that might be critical
print("\nNaN values per column in cleaned DataFrame:")
print(df_train_cleaned.isnull().sum()[df_train_cleaned.isnull().sum() > 0])
```

```python
# Drop 'TVT_input' due to a very high number of NaNs
df_train_processed = df_train_cleaned.drop(columns=['TVT_input']).copy()

# Impute missing values in feature columns with their median
# For 'GR', 'ANCC', and 'EGFDL'
for col in ['GR', 'ANCC', 'EGFDL']:
    if df_train_processed[col].isnull().sum() > 0:
        median_val = df_train_processed[col].median()
        # Using .loc to avoid FutureWarning with inplace=True on a chained assignment
        df_train_processed.loc[:, col] = df_train_processed[col].fillna(median_val)
        print(f"Missing values in '{col}' imputed with median: {median_val}")

print("\nNaN values per column after imputation:")
print(df_train_processed.isnull().sum()[df_train_processed.isnull().sum() > 0])

print("\nShape of DataFrame after processing NaNs:", df_train_processed.shape)
print("\nDataFrame head after processing:")
print(df_train_processed.head())
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Encode the 'Geology' column (target variable)
label_encoder = LabelEncoder()
df_train_processed['Geology_encoded'] = label_encoder.fit_transform(df_train_processed['Geology'])

print("Original Geology classes:", label_encoder.classes_)
print("Encoded Geology classes (first 5 mappings):")
for i, geo_class in enumerate(label_encoder.classes_[:5]):
    print(f"  {geo_class}: {i}")
```

```python
# Define features (X) and target (y)
# Drop 'Geology' (original string), 'Geology_encoded' (target), and 'well_id' (identifier)
# All other numerical columns are features

X = df_train_processed.drop(columns=['Geology', 'Geology_encoded', 'well_id'])
y = df_train_processed['Geology_encoded']

print(f"Features shape: {X.shape}")
print(f"Target shape: {y.shape}")
print("\nFeatures (X) columns:", X.columns.tolist())
```

```python
# Identify classes with only one member
class_counts = y.value_counts()
single_member_classes = class_counts[class_counts < 2].index

# Filter out rows corresponding to single-member classes
# First, identify the indices to keep
indices_to_keep = ~y.isin(single_member_classes)

X_filtered = X[indices_to_keep]
y_filtered = y[indices_to_keep]

print(f"Original number of samples: {len(y)}")
print(f"Number of single-member classes removed: {len(single_member_classes)}")
print(f"Number of samples after filtering: {len(y_filtered)}")

# Split the filtered data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_filtered, y_filtered, test_size=0.2, random_state=42, stratify=y_filtered)

print(f"\nX_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
# Scale numerical features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Features scaled successfully.")
print(f"X_train_scaled shape: {X_train_scaled.shape}")
print(f"X_test_scaled shape: {X_test_scaled.shape}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Initialize the Random Forest Classifier
# n_estimators: number of trees in the forest
# random_state: for reproducibility
# Consider adjusting n_estimators and other hyperparameters based on performance and computational resources
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1) # n_jobs=-1 uses all available cores

print("Training Random Forest Classifier...")
# Train the model on the scaled training data
rf_classifier.fit(X_train_scaled, y_train)
print("Random Forest Classifier training complete.")
```

```python
print("Evaluating Random Forest Classifier...")
# Predict on the scaled test data
y_pred = rf_classifier.predict(X_test_scaled)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"\nRandom Forest Classifier Accuracy: {accuracy:.4f}")

# Display classification report
# To handle the ValueError, we need to ensure target_names match the actual labels present in y_test
# We get the unique labels from y_test and map them back to their original names
unique_labels_in_test = sorted(y_test.unique())
actual_target_names = [label_encoder.inverse_transform([label])[0] for label in unique_labels_in_test]

print("\nClassification Report:")
print(classification_report(y_test, y_pred, labels=unique_labels_in_test, target_names=actual_target_names))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Helper function to extract geology layer boundaries from a single well's data
def get_geology_layers(df_well, label_encoder):
    layers = []
    current_geology_encoded = None
    start_tvt = None

    # Sort by TVT to ensure correct order
    df_well_sorted = df_well.sort_values(by='TVT', ascending=True).reset_index(drop=True)

    # Iterate through the sorted data to identify continuous geology segments
    for index, row in df_well_sorted.iterrows():
        tvt = row['TVT']
        geology_encoded = row['Geology_encoded']

        if current_geology_encoded is None:
            # Initialize with the first row's geology and TVT
            current_geology_encoded = geology_encoded
            start_tvt = tvt
        elif geology_encoded != current_geology_encoded:
            # Geology changed, record the previous layer
            layers.append({
                'Geology_encoded': current_geology_encoded,
                'Geology': label_encoder.inverse_transform([current_geology_encoded])[0],
                'Start_TVT': start_tvt,
                'End_TVT': df_well_sorted.loc[index - 1, 'TVT']
            })
            # Start new layer tracking
            current_geology_encoded = geology_encoded
            start_tvt = tvt

    # Add the last layer after the loop finishes
    if current_geology_encoded is not None:
        layers.append({
            'Geology_encoded': current_geology_encoded,
            'Geology': label_encoder.inverse_transform([current_geology_encoded])[0],
            'Start_TVT': start_tvt,
            'End_TVT': df_well_sorted.loc[df_well_sorted.index[-1], 'TVT']
        })

    return pd.DataFrame(layers)

# 1. Select a specific well_id for visualization
# Using the second unique well from the processed data for a new visualization
sample_well_id = df_train_processed['well_id'].unique()[1]
print(f"Visualizing geology layers for Well ID: {sample_well_id}")

# 2. Filter df_train_processed for that well
df_single_well = df_train_processed[df_train_processed['well_id'] == sample_well_id].copy()

# 3. Extract geology layers using the helper function
geology_layers_df = get_geology_layers(df_single_well, label_encoder)

# Ensure the dataframe is not empty before plotting
if geology_layers_df.empty:
    print(f"No geology layers found for well {sample_well_id}.")
else:
    # 4. Create a stratigraphic column-like plot
    plt.figure(figsize=(8, 12))

    # Map geology types to colors for consistency
    unique_geologies = geology_layers_df['Geology'].unique()
    # Using a categorical palette to ensure distinct colors for different geologies
    color_map = sns.color_palette('tab20', n_colors=len(unique_geologies))
    geology_color_dict = {geo: color_map[i % len(color_map)] for i, geo in enumerate(unique_geologies)}

    # Draw rectangles for each geology layer
    for index, row in geology_layers_df.iterrows():
        geology_type = row['Geology']
        start_tvt = row['Start_TVT']
        end_tvt = row['End_TVT']

        # Use a dummy x-axis for the column width (e.g., from 0 to 1)
        plt.fill_between([0, 1], start_tvt, end_tvt, color=geology_color_dict[geology_type], edgecolor='black', linewidth=0.5)

        # Add text label in the middle of the layer if the layer is thick enough
        mid_tvt = (start_tvt + end_tvt) / 2
        layer_thickness = end_tvt - start_tvt
        if layer_thickness > (df_single_well['TVT'].max() - df_single_well['TVT'].min()) * 0.01: # Label if layer is at least 1% of total depth range
            plt.text(0.5, mid_tvt, geology_type,
                     horizontalalignment='center', verticalalignment='center',
                     color='black', fontsize=9, weight='bold', bbox=dict(facecolor='white', alpha=0.7, edgecolor='none', boxstyle='round,pad=0.2'))

    plt.gca().invert_yaxis() # Depth increases downwards

    # Set y-limits with some padding
    min_tvt = geology_layers_df['Start_TVT'].min()
    max_tvt = geology_layers_df['End_TVT'].max()
    padding = (max_tvt - min_tvt) * 0.05
    plt.ylim(max_tvt + padding, min_tvt - padding)

    plt.title(f'Geology Layers for Well: {sample_well_id}')
    plt.ylabel('True Vertical Depth (TVT)')
    plt.xticks([]) # Hide x-axis ticks as it's a dummy axis for the column
    plt.xlabel('') # Hide x-axis label

    # Create a custom legend manually, to avoid duplicate labels from fill_between
    legend_handles = []
    for geo_type in unique_geologies:
        legend_handles.append(plt.Line2D([0], [0], color=geology_color_dict[geo_type], lw=10, label=geo_type))
    plt.legend(handles=legend_handles, title='Geology Types', bbox_to_anchor=(1.05, 1), loc='upper left')

    plt.tight_layout()
    plt.show()
```

### Create Submission File

```python
import pandas as pd
import os

test_directory_path = '/content/rogii-wellbore-geology-prediction/test'
submission_df_list = []

# Identify feature columns (excluding 'Geology', 'Geology_encoded', 'well_id')
feature_cols = X.columns.tolist()

# Calculate medians for ALL feature columns from the training set (X)
# This will be used to fill completely missing columns in test data
train_feature_medians = df_train_processed[feature_cols].median().to_dict()

print("Processing test wells...")
for filename in os.listdir(test_directory_path):
    if filename.endswith('__horizontal_well.csv'):
        well_id = filename.split('__')[0]
        file_path = os.path.join(test_directory_path, filename)
        df_test_well = pd.read_csv(file_path)

        # Handle the TVT column: If 'TVT' is missing but 'TVT_input' exists, rename it.
        if 'TVT' not in df_test_well.columns and 'TVT_input' in df_test_well.columns:
            df_test_well.rename(columns={'TVT_input': 'TVT'}, inplace=True)
            print(f"Renamed 'TVT_input' to 'TVT' for well {well_id}")
        elif 'TVT' not in df_test_well.columns:
            print(f"Warning: Neither 'TVT' nor 'TVT_input' column found for well {well_id}. Skipping.")
            continue

        # Store original TVT for submission file BEFORE any potential imputation changes
        original_tvt = df_test_well['TVT'].copy()

        # Ensure all feature columns from training are present in the test well dataframe
        # If a column is missing, add it and fill with the median from training data
        for col in feature_cols:
            if col not in df_test_well.columns:
                df_test_well[col] = train_feature_medians.get(col, 0) # Use training median, default to 0 if not found

        # Impute missing values in feature columns using medians from training data
        # This handles NaNs within existing columns and newly added columns
        for col in feature_cols:
            if df_test_well[col].isnull().any():
                df_test_well[col] = df_test_well[col].fillna(train_feature_medians.get(col, 0))

        # Select only the feature columns that the model was trained on, ensuring correct order
        X_test_well = df_test_well[feature_cols]

        # Scale features using the *fitted* scaler from training data
        X_test_well_scaled = scaler.transform(X_test_well)

        # Make predictions
        predictions_encoded = rf_classifier.predict(X_test_well_scaled)

        # Inverse transform predictions to original geology names
        predictions_geology = label_encoder.inverse_transform(predictions_encoded)

        # Create a DataFrame for this well's predictions in submission format
        well_submission_df = pd.DataFrame({
            'well_id': well_id,
            'TVT': original_tvt, # Use original TVT for submission
            'Geology': predictions_geology
        })
        submission_df_list.append(well_submission_df)

if submission_df_list:
    final_submission_df = pd.concat(submission_df_list, ignore_index=True)
    print(f"Generated submission DataFrame with shape: {final_submission_df.shape}")
    print("First 5 rows of submission DataFrame:")
    display(final_submission_df.head())

    # Save to CSV
    submission_file_path = '/content/submission.csv'
    final_submission_df.to_csv(submission_file_path, index=False)
    print(f"Submission file saved to {submission_file_path}")
else:
    print("No test horizontal_well.csv files found to process or all were skipped.")
```

### Analyzing Model Performance with respect to TVT (True Vertical Depth)

Although the model demonstrated very high overall accuracy, let's investigate any potential misclassifications and how they relate to the `TVT` feature. We'll extract the `TVT` values from the test set along with the true and predicted geology labels to understand if certain depth ranges are more prone to errors.

### Developing a Regression Model to Predict True Vertical Depth (TVT)

Now, let's build a regression model. We will use `TVT` (True Vertical Depth) as our continuous target variable and predict it based on the other available features. We will use a `RandomForestRegressor` for this task.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Define the new target variable (y_reg) as TVT
y_reg = df_train_processed['TVT']

# Define features (X_reg) by dropping TVT and categorical/identifier columns
# We will keep GR as a feature for predicting TVT
drop_cols = ['TVT', 'Geology', 'well_id', 'Geology_encoded']
X_reg = df_train_processed.drop(columns=drop_cols)

print(f"Regression Features shape: {X_reg.shape}")
print(f"Regression Target shape: {y_reg.shape}")
print("\nRegression Features (X_reg) columns:", X_reg.columns.tolist())

# --- Start: Data Sampling for faster regression training ---
sample_fraction = 0.1 # Adjust this value to control the sample size (e.g., 0.1 for 10%)
print(f"\nSampling {sample_fraction*100:.0f}% of the data for regression training...")

X_reg_sampled, _, y_reg_sampled, _ = train_test_split(X_reg, y_reg, test_size=(1-sample_fraction), random_state=42)

# Use the sampled data for the train-test split
# Split the data into training and testing sets for regression
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg_sampled, y_reg_sampled, test_size=0.2, random_state=42
)
# --- End: Data Sampling for faster regression training ---

print(f"\nX_train_reg shape: {X_train_reg.shape}")
print(f"X_test_reg shape: {X_test_reg.shape}")
print(f"y_train_reg shape: {y_train_reg.shape}")
print(f"y_test_reg shape: {y_test_reg.shape}")

# Scale numerical features for regression
scaler_reg = StandardScaler()
X_train_reg_scaled = scaler_reg.fit_transform(X_train_reg)
X_test_reg_scaled = scaler_reg.transform(X_test_reg)

print("\nRegression features scaled successfully.")

# Initialize and train the Random Forest Regressor
print("\nTraining Random Forest Regressor...")
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
rf_regressor.fit(X_train_reg_scaled, y_train_reg)
print("Random Forest Regressor training complete.")

# Make predictions on the test set
y_pred_reg = rf_regressor.predict(X_test_reg_scaled)

# Evaluate the regression model
mse = mean_squared_error(y_test_reg, y_pred_reg)
rmse = np.sqrt(mse)
r2 = r2_score(y_test_reg, y_pred_reg)

print(f"\nMean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R-squared (R2): {r2:.4f}")
```

### Visualizing Regression Results: Actual vs. Predicted TVT

This scatter plot compares the actual True Vertical Depth (TVT) values from the test set against the TVT values predicted by our Random Forest Regressor. A perfect model would show all points lying perfectly on the red diagonal line (where `Actual TVT = Predicted TVT`). Deviations from this line indicate prediction errors.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 7))
sns.scatterplot(x=y_test_reg, y=y_pred_reg, alpha=0.3)

# Add a perfect prediction line (y=x)
plt.plot([y_test_reg.min(), y_test_reg.max()], [y_test_reg.min(), y_test_reg.max()],
         color='red', linestyle='--', linewidth=2, label='Perfect Prediction')

plt.title('Actual vs. Predicted True Vertical Depth (TVT) Values')
plt.xlabel('Actual TVT')
plt.ylabel('Predicted TVT')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

### Create Submission File for Entire Test Dataset

```python
import pandas as pd
import os
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import shutil
from google.colab import drive

# --- Ensure Google Drive is mounted and data is copied for self-contained execution ---

# Mount Google Drive if not already mounted
if not os.path.exists('/content/drive'):
    print("Mounting Google Drive...")
    drive.mount('/content/drive')
else:
    print("Google Drive already mounted.")

source_path_data = '/content/drive/MyDrive/Wellbore_Geology/rogii-wellbore-geology-prediction'
destination_path_data = '/content/rogii-wellbore-geology-prediction'

# Check if data directory already exists in /content/, if not, copy it
if not os.path.exists(destination_path_data) or not os.path.exists(os.path.join(destination_path_data, 'train')):
    print(f"Copying data from '{source_path_data}' to '{destination_path_data}'...")
    if os.path.exists(destination_path_data):
        shutil.rmtree(destination_path_data) # Remove incomplete/old directory if it exists
    os.makedirs(os.path.dirname(destination_path_data), exist_ok=True)
    shutil.copytree(source_path_data, destination_path_data)
    print("Data copying complete.")
else:
    print("Data directory already exists in /content/. Skipping copy.")

# --- End of data setup ---


# --- Re-define df_train, df_train_cleaned, df_train_processed for self-contained execution ---
train_directory_path = '/content/rogii-wellbore-geology-prediction/train'

# Group files by well ID
well_data = {}
for filename in os.listdir(train_directory_path):
    if filename.endswith('.csv'):
        well_id = filename.split('__')[0]
        if well_id not in well_data:
            well_data[well_id] = {'horizontal_well': None, 'typewell': None}

        if '__horizontal_well.csv' in filename:
            well_data[well_id]['horizontal_well'] = os.path.join(train_directory_path, filename)
        elif '__typewell.csv' in filename:
            well_data[well_id]['typewell'] = os.path.join(train_directory_path, filename)

merged_dataframes = []
missing_files_count = 0

print("Re-merging data for classification model submission...")
for well_id, files in well_data.items():
    if files['horizontal_well'] and files['typewell']:
        df_horizontal = pd.read_csv(files['horizontal_well'])
        df_typewell = pd.read_csv(files['typewell'])

        df_horizontal.sort_values(by='TVT', inplace=True)
        df_typewell.sort_values(by='TVT', inplace=True)

        merged_df = pd.merge_asof(
            df_horizontal,
            df_typewell[['TVT', 'Geology']],
            on='TVT',
            direction='nearest',
            suffixes=('_well', '_geology')
        )

        merged_df['well_id'] = well_id
        merged_dataframes.append(merged_df)
    else:
        missing_files_count += 1

if merged_dataframes:
    df_train = pd.concat(merged_dataframes, ignore_index=True)
    print(f"Re-merged training DataFrame shape: {df_train.shape}")
else:
    print("No dataframes were merged during re-processing.")
    raise ValueError("No training data could be re-merged. Cannot proceed with submission file generation.")

# Drop rows where 'Geology' is NaN, as these cannot be used for classification
df_train_cleaned = df_train.dropna(subset=['Geology']).copy()
print(f"Shape of DataFrame after dropping NaN Geology rows (re-processed): {df_train_cleaned.shape}")

# Drop 'TVT_input' due to a very high number of NaNs
df_train_processed = df_train_cleaned.drop(columns=['TVT_input']).copy()

# Impute missing values in feature columns with their median
for col in ['GR', 'ANCC', 'EGFDL']:
    if df_train_processed[col].isnull().sum() > 0:
        median_val = df_train_processed[col].median()
        df_train_processed.loc[:, col] = df_train_processed[col].fillna(median_val)
print("Missing values imputed in df_train_processed (re-processed).")
# --- End of re-defining df_train_processed ---


# Re-define label_encoder (from cell e7239347)
label_encoder = LabelEncoder()
df_train_processed['Geology_encoded'] = label_encoder.fit_transform(df_train_processed['Geology'])

# Re-define X (features) and y (target) for classification (from cell 8e13f252)
X = df_train_processed.drop(columns=['Geology', 'Geology_encoded', 'well_id'])
y = df_train_processed['Geology_encoded']

# Filter out single-member classes as done in cell 8614e5c7 for robust stratification
class_counts = y.value_counts()
single_member_classes = class_counts[class_counts < 2].index
indices_to_keep = ~y.isin(single_member_classes)
X_filtered = X[indices_to_keep]
y_filtered = y[indices_to_keep]

# Split the filtered data into training and testing sets (from cell 8614e5c7)
X_train, X_test, y_train, y_test = train_test_split(X_filtered, y_filtered, test_size=0.2, random_state=42, stratify=y_filtered)

# Re-define and fit scaler (from cell 18324ba3)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
# No need for X_test_scaled here, as we only need the fitted scaler

# Re-initialize and train the Random Forest Classifier (from cell 8b003707)
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
print("Re-training Random Forest Classifier for submission file generation...")
rf_classifier.fit(X_train_scaled, y_train)
print("Random Forest Classifier re-training complete.")
# --- End of ensuring necessary variables ---


test_directory_path = '/content/rogii-wellbore-geology-prediction/test'
submission_df_list = []

# Identify feature columns (excluding 'Geology', 'Geology_encoded', 'well_id')
feature_cols = X.columns.tolist()

# Calculate medians for ALL feature columns from the training set (X)
train_feature_medians = df_train_processed[feature_cols].median().to_dict()

print("Processing test wells for submission file...")
for filename in os.listdir(test_directory_path):
    if filename.endswith('__horizontal_well.csv'):
        well_id = filename.split('__')[0]
        file_path = os.path.join(test_directory_path, filename)
        df_test_well = pd.read_csv(file_path)

        # Handle the TVT column: If 'TVT' is missing but 'TVT_input' exists, rename it.
        if 'TVT' not in df_test_well.columns and 'TVT_input' in df_test_well.columns:
            df_test_well.rename(columns={'TVT_input': 'TVT'}, inplace=True)
        elif 'TVT' not in df_test_well.columns:
            print(f"Warning: Neither 'TVT' nor 'TVT_input' column found for well {well_id}. Skipping.")
            continue

        # Store original TVT for submission file BEFORE any potential imputation changes
        original_tvt = df_test_well['TVT'].copy()

        # Ensure all feature columns from training are present in the test well dataframe
        for col in feature_cols:
            if col not in df_test_well.columns:
                df_test_well[col] = train_feature_medians.get(col, 0) # Use training median, default to 0 if not found

        # Impute missing values in feature columns using medians from training data
        for col in feature_cols:
            if df_test_well[col].isnull().any():
                df_test_well[col] = df_test_well[col].fillna(train_feature_medians.get(col, 0))

        # Select only the feature columns that the model was trained on, ensuring correct order
        X_test_well = df_test_well[feature_cols]

        # Scale features using the *fitted* scaler from training data
        X_test_well_scaled = scaler.transform(X_test_well)

        # Make predictions using the Random Forest Classifier
        predictions_encoded = rf_classifier.predict(X_test_well_scaled)

        # Inverse transform predictions to original geology names
        predictions_geology = label_encoder.inverse_transform(predictions_encoded)

        # Create a DataFrame for this well's predictions in submission format
        well_submission_df = pd.DataFrame({
            'well_id': well_id,
            'TVT': original_tvt, # Use original TVT for submission
            'Geology': predictions_geology
        })
        submission_df_list.append(well_submission_df)

if submission_df_list:
    final_submission_df = pd.concat(submission_df_list, ignore_index=True)
    print(f"Generated submission DataFrame with shape: {final_submission_df.shape}")
    print("First 5 rows of submission DataFrame:")
    print(final_submission_df.head())

    # Save to CSV in Google Drive
    drive_submission_dir = '/content/drive/MyDrive/Wellbore_Geology/submissions/'
    os.makedirs(drive_submission_dir, exist_ok=True)
    submission_file_path = os.path.join(drive_submission_dir, 'submission.csv')
    final_submission_df.to_csv(submission_file_path, index=False)
    print(f"Submission file saved to {submission_file_path}")
else:
    print("No test horizontal_well.csv files found to process or all were skipped.")
```
