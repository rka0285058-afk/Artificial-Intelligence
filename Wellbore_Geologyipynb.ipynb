```python
import pandas as pd
import os

train_directory_path = '/content/rogii-wellbore-geology-prediction/train'

# Group files by well ID
well_data = {}
for filename in os.listdir(train_directory_path):
    if filename.endswith('.csv'):
        # Extract well ID (e.g., 'c99f9fae' from 'c99f9fae__horizontal_well.csv')
        well_id = filename.split('__')[0]
        if well_id not in well_data:
            well_data[well_id] = {'horizontal_well': None, 'typewell': None}

        if '__horizontal_well.csv' in filename:
            well_data[well_id]['horizontal_well'] = os.path.join(train_directory_path, filename)
        elif '__typewell.csv' in filename:
            well_data[well_id]['typewell'] = os.path.join(train_directory_path, filename)


# Merge data for each well
merged_dataframes = []
missing_files_count = 0

print("Merging data for each well...")
for well_id, files in well_data.items():
    if files['horizontal_well'] and files['typewell']:
        df_horizontal = pd.read_csv(files['horizontal_well'])
        df_typewell = pd.read_csv(files['typewell'])

        # Ensure 'TVT' column is present for merging, rename 'TVT_input' to 'TVT' if needed
        # Based on previous inspection, 'TVT' is already the primary TVD column.
        # This line is removed as 'TVT_input' had many NaNs and 'TVT' is complete.
        # if 'TVT_input' in df_horizontal.columns and 'TVT' not in df_horizontal.columns:
        #     df_horizontal.rename(columns={'TVT_input': 'TVT'}, inplace=True)

        # Sort both dataframes by TVT for merge_asof
        df_horizontal.sort_values(by='TVT', inplace=True)
        df_typewell.sort_values(by='TVT', inplace=True)

        # Use merge_asof to merge on the nearest TVT for geology labels
        # We only need 'TVT' and 'Geology' from df_typewell for the merge, other columns are features.
        merged_df = pd.merge_asof(
            df_horizontal,
            df_typewell[['TVT', 'Geology']],
            on='TVT',
            direction='nearest', # find the closest TVT in df_typewell
            suffixes=('_well', '_geology') # Suffixes are not strictly needed here as we drop GR from df_typewell
        )

        merged_df['well_id'] = well_id # Add well_id column for identification
        merged_dataframes.append(merged_df)
    else:
        missing_files_count += 1
        # print(f"Skipping well {well_id} due to missing horizontal_well or typewell file.")

print(f"Successfully merged data for {len(merged_dataframes)} wells.")
if missing_files_count > 0:
    print(f"Skipped {missing_files_count} wells due to missing files.")

# Concatenate all merged dataframes into a single training DataFrame
if merged_dataframes:
    df_train = pd.concat(merged_dataframes, ignore_index=True)
    print("\nCombined training DataFrame shape:", df_train.shape)
    print("\nCombined training DataFrame head:")
    print(df_train.head())
    print("\nCombined training DataFrame info:")
    df_train.info()
    print("\nUnique geology classes:", df_train['Geology'].unique())
else:
    print("No dataframes were merged.")
```

```python
print(f"Number of rows with missing Geology labels: {df_train['Geology'].isnull().sum()}")
print(f"Percentage of rows with missing Geology labels: {(df_train['Geology'].isnull().sum() / len(df_train)) * 100:.2f}%")

# Drop rows where 'Geology' is NaN, as these cannot be used for classification
df_train_cleaned = df_train.dropna(subset=['Geology']).copy()

print(f"\nShape of DataFrame after dropping NaN Geology rows: {df_train_cleaned.shape}")
print(f"Unique geology classes after cleaning: {df_train_cleaned['Geology'].unique()}")

# Also check for NaNs in feature columns that might be critical
print("\nNaN values per column in cleaned DataFrame:")
print(df_train_cleaned.isnull().sum()[df_train_cleaned.isnull().sum() > 0])
```

```python
# Drop 'TVT_input' due to a very high number of NaNs
df_train_processed = df_train_cleaned.drop(columns=['TVT_input']).copy()

# Impute missing values in feature columns with their median
# For 'GR', 'ANCC', and 'EGFDL'
for col in ['GR', 'ANCC', 'EGFDL']:
    if df_train_processed[col].isnull().sum() > 0:
        median_val = df_train_processed[col].median()
        # Using .loc to avoid FutureWarning with inplace=True on a chained assignment
        df_train_processed.loc[:, col] = df_train_processed[col].fillna(median_val)
        print(f"Missing values in '{col}' imputed with median: {median_val}")

print("\nNaN values per column after imputation:")
print(df_train_processed.isnull().sum()[df_train_processed.isnull().sum() > 0])

print("\nShape of DataFrame after processing NaNs:", df_train_processed.shape)
print("\nDataFrame head after processing:")
print(df_train_processed.head())
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Encode the 'Geology' column (target variable)
label_encoder = LabelEncoder()
df_train_processed['Geology_encoded'] = label_encoder.fit_transform(df_train_processed['Geology'])

print("Original Geology classes:", label_encoder.classes_)
print("Encoded Geology classes (first 5 mappings):")
for i, geo_class in enumerate(label_encoder.classes_[:5]):
    print(f"  {geo_class}: {i}")
```

```python
# Define features (X) and target (y)
# Drop 'Geology' (original string), 'Geology_encoded' (target), and 'well_id' (identifier)
# All other numerical columns are features

X = df_train_processed.drop(columns=['Geology', 'Geology_encoded', 'well_id'])
y = df_train_processed['Geology_encoded']

print(f"Features shape: {X.shape}")
print(f"Target shape: {y.shape}")
print("\nFeatures (X) columns:", X.columns.tolist())
```

```python
# Identify classes with only one member
class_counts = y.value_counts()
single_member_classes = class_counts[class_counts < 2].index

# Filter out rows corresponding to single-member classes
# First, identify the indices to keep
indices_to_keep = ~y.isin(single_member_classes)

X_filtered = X[indices_to_keep]
y_filtered = y[indices_to_keep]

print(f"Original number of samples: {len(y)}")
print(f"Number of single-member classes removed: {len(single_member_classes)}")
print(f"Number of samples after filtering: {len(y_filtered)}")

# Split the filtered data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_filtered, y_filtered, test_size=0.2, random_state=42, stratify=y_filtered)

print(f"\nX_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
# Scale numerical features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Features scaled successfully.")
print(f"X_train_scaled shape: {X_train_scaled.shape}")
print(f"X_test_scaled shape: {X_test_scaled.shape}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Initialize the Random Forest Classifier
# n_estimators: number of trees in the forest
# random_state: for reproducibility
# Consider adjusting n_estimators and other hyperparameters based on performance and computational resources
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1) # n_jobs=-1 uses all available cores

print("Training Random Forest Classifier...")
# Train the model on the scaled training data
rf_classifier.fit(X_train_scaled, y_train)
print("Random Forest Classifier training complete.")
```

```python
print("Evaluating Random Forest Classifier...")
# Predict on the scaled test data
y_pred = rf_classifier.predict(X_test_scaled)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"\nRandom Forest Classifier Accuracy: {accuracy:.4f}")

# Display classification report
# To handle the ValueError, we need to ensure target_names match the actual labels present in y_test
# We get the unique labels from y_test and map them back to their original names
unique_labels_in_test = sorted(y_test.unique())
actual_target_names = [label_encoder.inverse_transform([label])[0] for label in unique_labels_in_test]

print("\nClassification Report:")
print(classification_report(y_test, y_pred, labels=unique_labels_in_test, target_names=actual_target_names))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Helper function to extract geology layer boundaries from a single well's data
def get_geology_layers(df_well, label_encoder):
    layers = []
    current_geology_encoded = None
    start_tvt = None

    # Sort by TVT to ensure correct order
    df_well_sorted = df_well.sort_values(by='TVT', ascending=True).reset_index(drop=True)

    # Iterate through the sorted data to identify continuous geology segments
    for index, row in df_well_sorted.iterrows():
        tvt = row['TVT']
        geology_encoded = row['Geology_encoded']

        if current_geology_encoded is None:
            # Initialize with the first row's geology and TVT
            current_geology_encoded = geology_encoded
            start_tvt = tvt
        elif geology_encoded != current_geology_encoded:
            # Geology changed, record the previous layer
            layers.append({
                'Geology_encoded': current_geology_encoded,
                'Geology': label_encoder.inverse_transform([current_geology_encoded])[0],
                'Start_TVT': start_tvt,
                'End_TVT': df_well_sorted.loc[index - 1, 'TVT']
            })
            # Start new layer tracking
            current_geology_encoded = geology_encoded
            start_tvt = tvt

    # Add the last layer after the loop finishes
    if current_geology_encoded is not None:
        layers.append({
            'Geology_encoded': current_geology_encoded,
            'Geology': label_encoder.inverse_transform([current_geology_encoded])[0],
            'Start_TVT': start_tvt,
            'End_TVT': df_well_sorted.loc[df_well_sorted.index[-1], 'TVT']
        })

    return pd.DataFrame(layers)

# 1. Select a specific well_id for visualization
# Using the second unique well from the processed data for a new visualization
sample_well_id = df_train_processed['well_id'].unique()[1]
print(f"Visualizing geology layers for Well ID: {sample_well_id}")

# 2. Filter df_train_processed for that well
df_single_well = df_train_processed[df_train_processed['well_id'] == sample_well_id].copy()

# 3. Extract geology layers using the helper function
geology_layers_df = get_geology_layers(df_single_well, label_encoder)

# Ensure the dataframe is not empty before plotting
if geology_layers_df.empty:
    print(f"No geology layers found for well {sample_well_id}.")
else:
    # 4. Create a stratigraphic column-like plot
    plt.figure(figsize=(8, 12))

    # Map geology types to colors for consistency
    unique_geologies = geology_layers_df['Geology'].unique()
    # Using a categorical palette to ensure distinct colors for different geologies
    color_map = sns.color_palette('tab20', n_colors=len(unique_geologies))
    geology_color_dict = {geo: color_map[i % len(color_map)] for i, geo in enumerate(unique_geologies)}

    # Draw rectangles for each geology layer
    for index, row in geology_layers_df.iterrows():
        geology_type = row['Geology']
        start_tvt = row['Start_TVT']
        end_tvt = row['End_TVT']

        # Use a dummy x-axis for the column width (e.g., from 0 to 1)
        plt.fill_between([0, 1], start_tvt, end_tvt, color=geology_color_dict[geology_type], edgecolor='black', linewidth=0.5)

        # Add text label in the middle of the layer if the layer is thick enough
        mid_tvt = (start_tvt + end_tvt) / 2
        layer_thickness = end_tvt - start_tvt
        if layer_thickness > (df_single_well['TVT'].max() - df_single_well['TVT'].min()) * 0.01: # Label if layer is at least 1% of total depth range
            plt.text(0.5, mid_tvt, geology_type,
                     horizontalalignment='center', verticalalignment='center',
                     color='black', fontsize=9, weight='bold', bbox=dict(facecolor='white', alpha=0.7, edgecolor='none', boxstyle='round,pad=0.2'))

    plt.gca().invert_yaxis() # Depth increases downwards

    # Set y-limits with some padding
    min_tvt = geology_layers_df['Start_TVT'].min()
    max_tvt = geology_layers_df['End_TVT'].max()
    padding = (max_tvt - min_tvt) * 0.05
    plt.ylim(max_tvt + padding, min_tvt - padding)

    plt.title(f'Geology Layers for Well: {sample_well_id}')
    plt.ylabel('True Vertical Depth (TVT)')
    plt.xticks([]) # Hide x-axis ticks as it's a dummy axis for the column
    plt.xlabel('') # Hide x-axis label

    # Create a custom legend manually, to avoid duplicate labels from fill_between
    legend_handles = []
    for geo_type in unique_geologies:
        legend_handles.append(plt.Line2D([0], [0], color=geology_color_dict[geo_type], lw=10, label=geo_type))
    plt.legend(handles=legend_handles, title='Geology Types', bbox_to_anchor=(1.05, 1), loc='upper left')

    plt.tight_layout()
    plt.show()
```
